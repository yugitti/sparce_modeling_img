```python
import cv2
import numpy as np
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import numpy as np
import random
%matplotlib inline
custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white'}
sns.set_style("dark", rc=custom_style)
plt.style.use('dark_background')
```

```python
img = cv2.imread("./blood.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
mu = np.mean(img_gray)
sigma = np.std(img_gray)
dB = 10
img_noise = img_gray + 10**(-dB/20)*np.reshape([random.gauss(mu, sigma) for i in range(np.size(img_gray))], np.shape(img_gray))
img_noise = (img_noise - img_noise.min()) / (img_noise.max() - img_noise.min()) * 255


img_noise = np.round(img_noise).astype(int)
img_noise[img_noise>255] = 255
img_noise[img_noise<0] = 0
```

```python
plt.gray()
figs, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].imshow(img_gray)
axs[1].imshow(img_noise)
```

# L1正則化

```python
lam = 30
alpha = 200
```

```python
b = img_noise - alpha
img_l1 = np.where(np.abs(b) - lam > 0, np.abs(b) - lam, 0) * np.sign(b) + alpha
```

```python
plt.gray()
figs, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].imshow(img_gray)
axs[1].imshow(img_noise)
axs[2].imshow(img_l1)
```
